Library Managemnet System

```python
books = ['Harry Potter', 'Atomic Habits']

print("Current list:", books)

new = input("Enter book name to add: ")
books.append(new)

search = input("Book you want to search: ")

if search in books:
    print("Book found at index:", books.index(search))
else:
    print("No book found")

rem = input("Enter book to remove: ")
books.remove(rem)

books.sort()
print("Sorted books:", books)
```
