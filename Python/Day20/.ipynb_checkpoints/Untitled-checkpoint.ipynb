```python
import numpy as np
```

```python
students = np.array(["Harsh", "Laukik", "Utkarsh", "Aditya", "Dishant","Shiva"])

marks = np.array([
    [85, 90, 78, 92, 88],
    [70, 75, 80, 72, 68],
    [95, 98, 94, 97, 96],
    [60, 65, 58, 62, 64],
    [88, 84, 91, 87, 90],
    [45,50,45,35,25]
])

print("Students:")
print(students)

print("\nMarks:")
print(marks)
```

```python
avg = np.mean(marks, axis=1)

print("\nAverage Marks:")
for i in range(len(students)):
    print(students[i], ":", avg[i])
```

```python
print("\nHighest and Lowest Marks:")
for i in range(len(students)):
    print(students[i], "Highest =", np.max(marks[i]), "Lowest =", np.min(marks[i]))
```

```python
print("\nStandard Deviation:")
for i in range(len(students)):
    print(students[i], ":", round(np.std(marks[i]), 2))
```

```python
rank = np.argsort(avg)[::-1]

print("\nRanking:")
for i in rank:
    print(students[i], ":", avg[i])
```

```python
class_avg = np.mean(avg)

print("\nClass Average =", class_avg)

print("\nStudents Above Class Average:")
for i in range(len(students)):
    if avg[i] > class_avg:
        print(students[i])
        
```

```python
subject_avg = np.mean(marks, axis=0)

print("\nSubject-wise Average:")
for i in range(len(subject_avg)):
    print("Subject", i + 1, ":", subject_avg[i])
```

```python
percentage = np.sum(marks, axis=1) / 5

print("\nPercentage:")
for i in range(len(students)):
    print(students[i], ":", percentage[i], "%")
```

```python
labels = students
values = percentage

print("\nVisualization Ready Data")
print("Labels:", labels)
print("Values:", values)
```
