```python
for i in range(1,3):
    for j in range(1,3):
        for k in range(1,3):
            print(i,j,k)
```

```python
for i in range(0,3):
    for j in range(0,3):
        for k in range(0,3):
            if((i+j+k)%2 == 0):
                print(i,j,k)
```

```python
for i in range(0,3):
    for j in range(0,3):
        for k in range(0,3):
            if(i == j == k):
                print(i,j,k)
```

```python
for i in range(0,3):
    for j in range(0,3):
        for k in range(0,3):
            if(i != j != k != i):
                print(i,j,k)
```
