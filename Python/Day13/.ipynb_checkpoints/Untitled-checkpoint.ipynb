**Dictionary**

```python
student = {
    "Name" : 'Rahul',
    "Age" : 22,
    "CGPA" : 8.4
}

student["city"] = "Delhi"

student.pop("city")
print(student)

student.popitem()
print(student)
```

```python
print(student['city'])
print(student.get("Name"))
print(student.get("school"))
```

```python
square = {x:x**2 for x in range(1,6)}
print(square)

even_square = {x:x**2 for x in range(1,11) if x%2 == 0}
print(even_square)
```

```python
student = {"name" : "bheem","age" : 20}
print(student)
student['school'] = 'rvce'
print(student)
student["name"] = "ram"
print(student)
student.pop('school')
print(student)
for i,j in student.items():
    print(f"{i} : {j}")
```

```python
students = {
    "stu1" : {"name" : "ram","age" : 20},
    "stu2" : {"name" : "shaam","age" : 22}
}
for i,j in students.items():
    print(f"{i} : {j}")

    
```

```python
my_string = "hello world"

frequency = {char: my_string.count(char) for char in set(my_string)}

print(frequency)
```
