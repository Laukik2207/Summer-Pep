```python
from functools import reduce

numbers = [1,2,20,5]
largest = reduce(lambda a,b :a  if a > b else b,numbers)
print(largest)
```

```python
def prime(x):
    if(x <=1 ):
        return -1
    i = 2
    for i in range(x):
        if(x%2 == 0):
            return 0
    else:
        return 1

numbers = [2,37,5,6,5,7]

is_prime = list(filter(prime,numbers))
print(is_prime)
```
