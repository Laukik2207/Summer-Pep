**Nested Loops**

```python
#1
for i in range(3):
    for j in range(3):
        print(i,j)
```

```python
#2
for i in range(3):
    print("Row ",i)
    for j in range(3):
        print(" Column ",j)
```

```python
#3
for i in range(3):
    for j in range(4):
        print("* ",end = " ")
    print()
```

```python
#4
for i in range(3):
    for j in range(1,5):
        print(j,end=" ")
    print()
```

```python
#5
for i in range(1,4):
    for j in range(1,4):
        print(i*j,end=" ")
    print()
```

```python
#6
k=1
for i in range(3):
    for j in range(3):
        print(k,end=" ")
        k+=1
    print()
```

```python
#7
k = 1
for i in range(3):
    for j in range(3):
        print(2*k,end=" ")
        k+=1
    print()
            
```
