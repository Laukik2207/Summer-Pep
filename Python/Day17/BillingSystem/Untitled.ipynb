```python
import os

project = "billingSystem"

# Create project folder
os.makedirs(project, exist_ok=True)

# Create Python files
files = [
    "main.py",
    "calculator.py",
    "tax.py",
    "discount.py",
    "invoice.py"
]

for file in files:
    open(os.path.join(project, file), "w").close()

print("Project created successfully!")
```

```python
import os

project = "billingSystem"

os.makedirs(project, exist_ok=True)

contents = {
    "main.py": "# Main program\n",
    "calculator.py": "# Calculator functions\n",
    "tax.py": "# Tax calculation functions\n",
    "discount.py": "# Discount calculation functions\n",
    "invoice.py": "# Invoice generation functions\n"
}

for filename, content in contents.items():
    with open(os.path.join(project, filename), "w") as f:
        f.write(content)

print("Billing System project created successfully!")
```
