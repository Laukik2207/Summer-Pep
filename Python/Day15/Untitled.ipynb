**File Handling**

```python
with open("file.txt","w") as file:
    file.write("101 , Rahul,50000\n")
    file.write("102 , Priya,60000\n")

with open("file.txt","r") as file:
    print(file.read())



with open("file.txt","r") as file:
    print(file.read(5))
    file.seek(10)
    print(file.read(10))
    file.seek(0)
    print(file.readline())
    print(file.readlines())
```

```python
with open("intro.txt","x") as file:
    file.write("Helloo! I am Laukik\n")
    print("File created")
    file.close()
```

```python
with open("intro.txt","r") as file:
    print(file.read())

with open("intro.txt","a") as file:
    file.write("\nText Appended!")
    print("Text Appended")

with open("intro.txt","r") as file:
    lines = 0
    while file.readline() != "":
        lines += 1
    print(lines)
```

```python
with open("intro.txt", "r") as file:
    text = file.read()

words = len(text.split())
print("Number of words:", words)
```

```python
word = input("Enter the word to search: ")

with open("intro.txt", "r") as file:
    words = file.read().split()

if word in words:
    print("Word index:", words.index(word))
else:
    print("Word not found")
```

```python
ip = input("Enter")
nf = input("New file")

with open("ip","r") as file:
    with open("nf","x") as newfile:
        newfile.write(file.read())
        print("Copy successfull")
```
