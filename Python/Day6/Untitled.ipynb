```python
#calculate function
def calc(n1,n2=0,op="+"):
    if(op == 'sq'):
        print(n1*n2)
    elif(op == 'cube'):
        print(n1*n1*n1)
    else:
        r = str(n1)+str(op)+str(n2)
        print(eval(r))

calc(10,5,"+")
    
```

```python
#Larger function
def larger(n1,n2):
    if(n1>n2):
        print(n1,"is larger")
    else:
        print(n2,"is larger")

larger(10,15)
```

```python
#EvenOdd Function
def evenodd(n1):
    if(n1 % 2 == 0):
        print("Even")
    else:
        print("Odd")

n1 = int(input("Enter num 1:"))
evenodd(n1)
```

```python
#Larger From 5 Nums Function

def larger(n1,n2,n3,n4,n5):
    max = 0
    if(n1>n2):
        max = n1
    else:
        max = n2
    if( n3 > max):
        max = n3
    if(n4 > max):
        max = n4
    if(n5 > max):
        max = n5
    print(max)



larger(1,5,3,4,2)
```
