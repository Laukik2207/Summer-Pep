```python
def fact(n):
    if(n == 1):
        return 1 
    return n*fact(n-1)

n = int(input("Enter number: "))
print(fact(n))
```

```python
def sum(n):
    if(n == 1):
        return 1
    return n+sum(n-1)

sum(5)
    
```
