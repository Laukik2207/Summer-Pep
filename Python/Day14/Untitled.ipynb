```python
purchases = {
    "Alice": {"Laptop", "Mouse", "Keyboard"},
    "Bob": {"Laptop", "Mouse", "Monitor"},
    "Charlie": {"Laptop", "Keyboard", "Headphones"},
    "David": {"Mouse", "Monitor", "Keyboard"},
    "Eve": {"Laptop", "Mouse", "Keyboard", "Monitor"}
}

current_customer = "Alice"

current_products = purchases[current_customer]

recommendations = set()

for customer, products in purchases.items():
    if customer == current_customer:
        continue

    common_products = current_products & products

    if len(common_products) > 0:
        recommendations.update(products - current_products)

print("Current Customer:", current_customer)
print("Purchased Products:", current_products)
print("Recommended Products:", recommendations)
```

```python
reservations = ["A1", "A2", "A3", "A2", "B1", "B2", "A3"]

seen = set()
double_booked = set()

for seat in reservations:
    if seat in seen:
        double_booked.add(seat)
    else:
        seen.add(seat)

print("Double Booked Seats:", double_booked)
```
