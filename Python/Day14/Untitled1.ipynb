**TRY & Catch**

```python
a = int(input("Enter num1:"))
b = int(input("Enter num2:"))
try:
    print(a/b)
except ZeroDivisionError:
    print("Cannot divide with 0");
    
```

```python
try:
    age = int(input("Enter age: "))
    if age >= 18:
        print("You are eligible")
    else:
        print("You are not eligible")
except ValueError:
    print("Enter a numerical value")
```
