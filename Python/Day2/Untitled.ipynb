```python
pin = 2207
balance = 5000

print("Please Swap Your Card.")

tempPin = int(input("Enter your card pin: "))
if(tempPin == pin):
    withdraw = int(input("Enter the amount you want to withdraw: "))
    if(withdraw > balance):
        print("Withdraw amount exceeds your balance!")
    else:
        print("\nWithdrawal successfull!")
        print("Your remaining balance is: ",balance - withdraw)
else:
    print("\nIncorrect Pin")

print("\nKindly remove your card. Thankyou!")
```
